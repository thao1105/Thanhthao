```python
import random 
x = []
s = 0 
for i in range(1000):
    x.append(random.randint(-1000, 1000))
f  = input("Nhập tên tệp từ bàn phím : ")   
for i in x:
    if s < 9:
        fo = open(f, 'a')
        chuoikytu = str(i)+','
        fo.write(chuoikytu)
        fo.close()
        s = 1
    else:
        fo = open(f, 'a')
        fo.write(str(i)+"\n" )
        fo.close()
        s = 0
fo = open(f, 'r')
a = fo.readlines()
for i in a:
    print(i.replace(",", "\t"))
fo.close()
```
