```python
import numpy as np 
import random 
from math import sin,cos,log,sqrt,tan
A = np.random.randint(low=5, high=10,
size=15).reshape((3,5))
print("ma trận A :")
print(A)
x = np.array(A)
print("sin(A):")
print(np.round(np.sin(x),3))
print("cos(A):")
print(np.round(np.cos(x),3))
print("tan(A):")
print(np.round(np.tan(x),3))
print("căn bậc hai của A :")
print(np.round(np.sqrt(x),3))
print("log(A)")
print(np.round(np.log10(x),3))
```
