```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def ham_bac_3(a, b, c, d,x):
  f = a*x**3 - b*x**2 + c
  return f
def ham_bac_2(a, b, c, x):
   f = a*x**2 + b*x + c
   return f
def ham_bac_1(a, b,x):
   f = -a*x + b
   return f
x = np.linspace(start=-10.0,stop=10.0, num=1000)
y3 = ham_bac_3(2, 3, -5,9, x)
y2 = ham_bac_2(-2,-5,3, x)
y1 = ham_bac_1(-5,3,x)
fig, ax = plt.subplots()
ax.plot(x,y3,label=r'$y=2x^{3}-3x^{2}-5x+9$')
ax.plot(x,y2,label=r'$y=-2x^{2}-5x+3$')
ax.plot(x,y1,label=r'$y=-5x + 3$')
ax.set_title("Đồ thị phương trình bậc 1 , bậc 2 và bậc 3")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
x = np.linspace(start=-10.0,stop=10.0, num=1000)
y4 = np.cos(x/np.pi)
y5 = np.sin((np.pi/3)*x) 
fig, ax = plt.subplots()
ax.plot(x,y4,label=r"$y4 = cos (x/np.pi)$")
ax.plot(x,y5,label=r"$y5 = sin (np.pi/3)*x)$") 
ax.set_title("Đồ thị của cos(x/π) và sin (π/3*x)")
ax.legend()
plt.show()
```
