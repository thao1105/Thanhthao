```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Xây dựng hàm số
def ham_bac_2(a, b, c, x):
   f = a*x**2 + b*x + c
   return f
#sinh dữ liệu
x = np.linspace(start=-10.0,
stop=10.0, num=200)
y = ham_bac_2(2, 3, -5, x)
#Tạo đối tượng chứa hình vẽ -
fig, ax = plt.subplots()
#Vẽ đồ thị hàm số bậc 2 với các
ax.plot(x, y)
#Hiển thị ra màn hình
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def ham_bac_3(a, b, c, d,x):
  f = a*x**3 - b*x**2 + c*x - d
  return f
def ham_bac_2(a, b, c, x):
   f = a*x**2 + b*x + c
   return f
def ham_bac_1(a, b,x):
   f = -a*x + b
   return f
x = np.linspace(start=-10.0,stop=10.0, num=1000)
y3 = ham_bac_3(2, 3, -5,9, x)
y2 = ham_bac_2(-2,-5,3, x)
y1 = ham_bac_1(-5,3,x)
fig, ax = plt.subplots()
ax.plot(x,y3,label=r'$y=2x^{3}-3x^{2}-5x+9$')
ax.plot(x,y2,label=r'$y=-2x^{2}-5x+3$')
ax.plot(x,y1,label=r'$y=-5x + 3$')
ax.set_title("Đồ thị phương trình bậc 1 , bậc 2 và bậc 3")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
x2 = np.linspace(start=-10.0,stop=10.0, num=1000)
y4 = np.cos(x/np.pi)
y5 = np.sin((np.pi/3)*x) 
fig, ax = plt.subplots()
ax.plot(x2,y4,label=r"$y4 = cos (x/np.pi)$")
ax.plot(x2,y5,label=r"$y5 = sin (np.pi/3)*x)$") 
ax.set_title("Đồ thị của cos(x/π) và sin (π/3*x)")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
def ham_Rastrigin(x, y):
  z = x**2 + y**2 - np.cos(12*x) - np.cos(18*y)
  return z 
x = np.linspace(start=-2.0, stop=2.0,num=1000)
y = np.linspace(start=-1.0, stop=3.0,num=1000)
x, y = np.meshgrid(x,y)
z = ham_Rastrigin( x, y)
fig, ax = plt.subplots(subplot_kw={"projection":"3d"})
rosen_surf = ax.plot_surface(x, y, z, cmap=cm.gist_heat_r,linewidth=0, antialiased=False)
ax.set_zlim(-2,4)
fig.colorbar(rosen_surf, shrink=0.5,aspect=5)
plt.show()
```
